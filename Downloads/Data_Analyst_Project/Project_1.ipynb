```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/vinodhkumar/Downloads/New_York_City_data.csv')
df
```

```python
movingavg_city = df['CITY_AVG_TEMP'].rolling(10).mean()
movingavg_global = df['GLOBAL_DATA_AVG_TEMP'].rolling(10).mean()
```

```python
plt.plot(df['YEAR'],movingavg_city,label ='Newyork_City' )
plt.plot(df['YEAR'],movingavg_global, label= 'Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (°C)") 
plt.title("Newyork City Average Temperature")
plt.show()
```
